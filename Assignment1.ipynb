# Program 1-A:  Quadratic Equation

A quadratic equation is a polynomial equation in terms of a variable $x$ as follows:  $ax^2 + bx + c = 0$.  The formula for solving a quadratic equation is: 

\begin{equation*}
x = \frac{-b \pm \sqrt{b^2 - 4ac}}{2a}
\end{equation*}

This equation calculates what is called the *roots* of the quadratic equation.  There are two roots to each quadratic equation, one created when the $\pm$ symbol is $+$, the other when it is $-$.  Depending on the values of the coefficients, $a$, $b$ and $c$ the roots can be either real or complex numbers.  

Write a Python program that inputs the 3 coefficients, $a$, $b$ and $c$, of a quadratic equation, then calculates and outputs the two roots of the equation.   

*Requirement:*  This program must be completed using the quadratic equation, simple Python calculations and conditional expressions.  In particular, the Python **complex** numbers **module** can**not** be **used**.   

About the square root:  To calculate  a mathematical expression $variable = \sqrt{expression}$ use the Python statement `variable = (expression)**(0.5)`.
